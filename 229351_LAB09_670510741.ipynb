# Classical decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib as mpl

mpl.rc("figure", figsize=(12,6))

# Remove the row that was added to restore the monthly frequency
data = data.drop(data.index[-1])

# Set the frequency of the index
data.index.freq = 'MS'

result_add = seasonal_decompose(data['value'], model='additive')

result_add.plot();
```

```python
result_mul = seasonal_decompose(data['value'], model='multiplicative')

result_mul.plot();
```

### ปฏิบัติการครั้งที่ 7
1. สร้างโมเดลเพื่อการทำนายดังนี้
1.1 แบ่งข้อมูลออกเป็น 2 ส่วน
   - training set: วันที่ 1996-01-01 ถึง 2009-12-01
   - test set: วันที่ 2010-01-01 ถึง 2012-03-01  
1.2 แยกส่วนประกอบ $y_t=S_t+T_t+R_t$ บน training set
1.3 สร้าง time series ชุดใหม่ที่แสดงถึงทำนายค่าบน test set โดยนำค่า $T_t+R_t$ ของวันล่าสุดใน training set ที่มีค่า $T_t$ มาบวกกับแต่ละค่าใน $S_t$ จากวันที่ วันที่ 2010-01-01 ถึง 2012-03-01  
3. คำนวณ RMSE โดยใช้ฟังก์ชัน `rmse` ข้างล่าง
4. แสดงแผนภาพข้อมูล elecequip และค่าทำนายที่ได้

```python
# 1.1
train_data = data['value'][:'2009-12-01']
test_data = data['value']['2010-01-01':]

# 1.2
result_train = seasonal_decompose(train_data, model='additive')

# 1.3
last_trend_resid = result_train.trend.iloc[-1] + result_train.resid.iloc[-1]

# Get the seasonal component for the test set period by reindexing
seasonal_test = result_train.seasonal.reindex(test_data.index)

# Create the forecast by adding the last trend+resid from training to the seasonal component of the test set
forecast = last_trend_resid + seasonal_test
```

```python
def rmse(y_true,y_pred):
  # Inputs
  # y_true: actual value, y_pred: predicted values
  # Output: RMSE between y_true and y_pred
  return np.sqrt(np.mean(np.square(y_true-y_pred)))
```

```python
ax = data['value'].plot(marker='.', color='black', figsize=(20,8), label="Actual Data")
test_data.plot(ax=ax, marker='o', color='blue', linestyle='--', label="Test Data")
forecast.plot(ax=ax, marker='o', color='red', label="Forecast")

ax.set_title("Electric Equipment Production with Forecast")
ax.legend()
plt.show()
```
